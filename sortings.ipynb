# Bubble Sort


```python
def bubblesort(arr):
    n= len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j] > arr[j+1]:
                arr[j] , arr[j+1] = arr[j+1] , arr[j]
    
    return arr
        


#Driver Code
arr= [70, 20, 50, 60, 35, 47]
result= bubblesort(arr)
print("Array after using bubble sort is : " , result)
```

# Selection Sort

```python
def selectionSort(arr):
    n= len(arr)
    for i in range(n):
        min = i
        for j in range(i+1,n):
            if arr[j] < arr[min]:
                min = j
        arr[i], arr[min] =arr[min] , arr[i] 
    return arr
        


#Driver Code
arr= [70, 20, 50, 60, 35, 47]
result= selectionSort(arr)
print("Array after using selection sort is : " , result)
```
