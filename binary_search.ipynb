## ğŸ§© Problem 1: Integer Square Root (Using Binary Search)

Compute and return the **square root** of a given non-negative integer `x`.  
Since the return type is an integer, the **decimal digits are truncated**,  
and only the **integer part** of the result is returned.

---

### ğŸ§  Example 1
**Input:**  `4`

**Output:**  `2`

### ğŸ§  Example 2
**Input:**  `8`

**Output:**  `2`


**Explanation:**  
The square root of `8` is approximately `2.8284...`.  
Since we only take the integer part, the output is **2**.

---

### ğŸ’¡ Hint
Try to use a **modified binary search approach** for an optimized solution instead of linear search.

---

### âš™ï¸ Constraints
- `0 <= x <= 10^12`



```python
def square_root(x):
    s = 0
    e = x
    ans = -1

    while s <= e:
        mid = (s + e) // 2
        if mid * mid == x:
            return mid  
        elif mid * mid < x:
            ans = mid   
            s = mid + 1
        else:
            e = mid - 1

    return ans  


x = 8
result = square_root(x)
print("Result:", result)
```

## ğŸ§© Problem 2: First Bad Version

You are a **product manager** and currently leading a team to develop a new product.  
Unfortunately, the latest version of your product **fails the quality check**.  

Since each version is developed based on the previous one,  
all the versions **after a bad version are also bad**.

Your task is to find out the **first bad version**, which causes all the following ones to be bad.

---

### ğŸ§  Example
**Input:**  `[0, 0, 1, 1, 1]`

**Output:**  `3`

**Explanation:**  
Here, `0` indicates a **good version** and `1` indicates a **bad version**.  
The **first bad version** appears at **index 3** (1-based index).

---

### âš™ï¸ Constraints
- You can only call the function `isBadVersion(version)` which returns:
  - `True` if the version is bad  
  - `False` otherwise
- You should **minimize the number of calls** to `isBadVersion()`.

---

### ğŸ’¡ Expected Solution
Use **binary search** to efficiently find the first bad version.

---

### â±ï¸ Time Complexity
The time complexity of the binary search approach is:
    n log (n)

##### [Similiar Leetcode](https://leetcode.com/problems/first-bad-version/)

```python
def version_check(v):
    l = 0
    r = len(v)-1

    while l <= r:
        mid = (l+r)//2
        if v[mid] == 1:
            if v[mid-1] == 1:
                r = mid - 1
            else:
                return mid+1
        elif v[mid] != 1:
            l = mid + 1
    return -1



## Driver code    
versions = [0, 0, 1, 1, 1]
result = version_check(versions)
print(f"First faulty version found on index: {result}")
```
