#### This is for Ternary Search 

##### It is similiar to binary search but it devide the searching area in three parts intsead Tow like binary Search


```python
# Using only while
def ternary_search(arr, target):
    l = 0
    r = len(arr)-1

    while (l < r):
        m1 = l+(r-l)//3
        m2 = r-(r-l)//3
        if arr[m1] == target:
            return m1+1
        elif arr[m2] == target:
            return m2+1
        elif arr[m1] >= target:
            r = m1-1
        elif arr[m2] <= target:
            l = m2+1
        else:
            l = m1+1
            r = m2-1
    return -1


# Driver code
arr = [2,5,7,8,12,32,65,98,100,120,135,140]
target = 2
result = ternary_search(arr,target)
print(result)
```

```python
# Using Recursive call
def ternary_search(arr, l, r, target):
    if l > r:
            return -1

    m1 = l+(r-l)//3
    m2 = r-(r-l)//3
    if arr[m1] == target:
        return m1+1
    if arr[m2] == target:
        return m2+1
    if arr[m1] > target:
        return ternary_search(arr, l, m1-1, target)
    elif arr[m2] < target:
        return ternary_search(arr, m2+1, r, target)
    else:
        return ternary_search(arr, m1+1, m2-1, target)

# Driver code
arr = [2,5,7,8,12,32,65,98,100,120,135,140]
target = 12
l = 0
r = len(arr)-1
result = ternary_search(arr,l,r,target)
print(result)
```
